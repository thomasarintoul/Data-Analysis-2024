 # Conditional Statements

## Question 

You are driving too fast and a police officer stops you. Write a **function** to return one of 3 possible results: "No ticket", "Small ticket", or "Big Ticket". If your speed is 70mph or less, the result is "No Ticket". If speed is between 71 and 80mph inclusive, the result is "Small Ticket". If speed is 81mph or more, the result is "Big Ticket". Unless it is your birthday -- on your birthday, the police officer will let you go at 5mph higher in all cases.
<br> 

Tip: For your birthday, use True and False statements.
<br> 

Demonstrate your code works using print statements.

### Answer:

*Your answer here:*

```python
def speeding_ticket(speed, birthday):
    '''
    :param speed: recorded speed in miles per hour
    :param birthday: boolean
    :return: resulting ticket
    '''
    legal_lim = 70
    small_lim = 80

    if birthday == True:
        if speed <= legal_lim:
            return print("Your speed is {:.2f} mph. You are within the speed limit: No Ticket".format(speed))
        if legal_lim < speed <= (legal_lim + 5):
            return print("Your speed is {:.2f} mph but it's your birthday: No Ticket".format(speed))
        if (legal_lim + 5) < speed <= small_lim:
            return print("Your speed is {:.2f} mph: Small Ticket".format(speed))
        if small_lim < speed <= (small_lim + 5):
            return print("Your speed is {:.2f} mph but it's your birthday: Small Ticket".format(speed))
        elif speed > (small_lim + 5):
            return print("Your speed is {:.2f} mph and the officer doesn't care that it's your birthday: Big Ticket".format(speed))
    if birthday == False:
        if speed <= legal_lim:
            return print("Your speed = {:2f} mph. You are within the speed limit: No Ticket")
        elif legal_lim < speed <= small_lim:
            return print("Your speed = {:2f} mph: Small Ticket")
        elif speed >= small_lim:
            return print("Your speed = {:2f} mph: Big Ticket")

speeding_ticket(60, False)
speeding_ticket(70, False)
speeding_ticket(71, False)
speeding_ticket(75, False)
speeding_ticket(80, False)
speeding_ticket(85, False)
speeding_ticket(86, False)

speeding_ticket(60, True)
speeding_ticket(70, True)
speeding_ticket(71, True)
speeding_ticket(75, True)
speeding_ticket(80, True)
speeding_ticket(85, True)
speeding_ticket(86, True)
```
